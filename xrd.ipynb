```python
from pymatgen.io.cif import CifWriter
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
m = MPRester("hzUX5FwAxRKy1v6G")
structure = m.get_structure_by_material_id("mp-10778")
CifWriter(structure,0.1, write_magmoms=False).write_file("mp-10778.cif")
#Creation structure de base
```

```python
xrdc = XRDCalculator(1.54060)
```

```python
p = xrdc.get_pattern(structure)
#Creation modèle de diffraction
```

# Pics pour un rayonnement X incident CuKα avec λ=1.54060 Å

```python
print("Visualisation des différents pics:")
xrdc.show_plot(structure, annotate_peaks = False) 
```

## Premier pic

```python
print('2\u0398: ' + str(p.x[0]) + '°' )
print('Intensité: ' + str(p.y[0])+' %')
print('Indices hkl : ', p.hkls[0])
print('Valeur de d : ', p.d_hkls[0])
```

## Deuxième pic

```python
print('2\u0398: ' + str(p.x[1]) + '°' )
print('Intensité : ' + str(p.y[1])+' %')
print('Indices hkl : ' + str(p.hkls[1]))
```
